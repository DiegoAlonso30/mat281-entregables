
### Parte 1: Limpieza y preparación

1. Revisar y describir el dataset:

   * ¿Cuántas filas y columnas tiene?
   * ¿Qué tipos de datos hay?
   * ¿Cuántos valores nulos hay por columna?

2. Transformar la columna `date_added` a tipo fecha.

3. Crear columnas auxiliares con `assign`:

   * Año (`year_added`)
   * Mes (`month_added`)



```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/netflix_titles.csv')
print("Número de filas y columnas:", df.shape)
print("\nTipos de datos:\n", df.dtypes)
print("\nValores nulos por columna:\n", df.isnull().sum())
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.assign(
    year_added=df['date_added'].dt.year,
    month_added=df['date_added'].dt.month
)
display(df.head())
```

## Parte 2: Técnicas avanzadas de pandas

4. Utilizar `.loc` para seleccionar películas (`type == 'Movie'`) que fueron agregadas después del año 2018.

5. Utilizar `str.contains()` y `str.extract()`:

   * Filtrar títulos que contienen la palabra 'love' (sin distinguir mayúsculas/minúsculas).
   * Extraer la duración en minutos para las películas desde la columna `duration`.

6. Aplicar `explode()` sobre la columna `listed_in` para obtener una fila por cada género.

7. Obtener un top 10 de géneros más frecuentes utilizando `value_counts()`.

8. Aplicar `where()` y `mask()` para marcar las películas de más de 120 minutos como contenido largo en una nueva columna.

9. Utilizar `.loc` para filtrar películas que cumplen con:

   * Más de 100 minutos de duración.
   * Rating igual a `'R'`.
   * País igual a `'United States'`.

10. Utilizar `.style` para formatear visualmente el top 10 de películas más largas.

```python
movies_after_2018 = df.loc[(df['type'] == 'Movie') & (df['year_added'] > 2018)]
display(movies_after_2018.head())
love_titles = df[df['title'].str.contains('love', case=False, na=False)]
display(love_titles.head())
df['duration_minutes'] = df['duration'].str.extract('(\d+)').astype(float)
display(df[['duration', 'duration_minutes']].head())
exploded_df = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')
display(exploded_df.head())
top_genres = exploded_df['listed_in'].value_counts().head(10)
display(top_genres)
df['content_length'] = df['duration_minutes'].where(df['duration_minutes'] <= 120, 'Long')
df['content_length'] = df['content_length'].mask(df['duration_minutes'] > 120, 'Long')
display(df[['duration_minutes', 'content_length']].head())
filtered_movies = df.loc[(df['duration_minutes'] > 100) & (df['rating'] == 'R') & (df['country'] == 'United States')]
display(filtered_movies.head())
top_10_long_movies = df[df['type'] == 'Movie'].sort_values(by='duration_minutes', ascending=False).head(10)
display(top_10_long_movies[['title', 'duration_minutes']].style.background_gradient(subset=['duration_minutes']))
```

```python
# Pregunta 11: ¿Cuáles son las combinaciones más frecuentes de género y rating en el dataset?
# Se utiliza el exploded_df creado en la Parte 2
genre_rating_counts = exploded_df.value_counts(subset=["listed_in", "rating"])
display(genre_rating_counts.head(10))

# Pregunta 12: ¿Existen películas con el mismo nombre (title) pero con distinto año de lanzamiento (release_year)?
duplicate_titles = df[df.duplicated(subset=['title'], keep=False)].sort_values(by='title')
duplicate_titles_diff_year = duplicate_titles[duplicate_titles.duplicated(subset=['title', 'release_year']) == False].groupby('title').filter(lambda x: x['release_year'].nunique() > 1)
display(duplicate_titles_diff_year[['title', 'release_year']].drop_duplicates().sort_values(by='title'))

# Pregunta 13. ¿Cuántos títulos únicos hay en total en la columna title?
unique_titles_count = df['title'].nunique()
print("\nNúmero de títulos únicos:", unique_titles_count)
```
